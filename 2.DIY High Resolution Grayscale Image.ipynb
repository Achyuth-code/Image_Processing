# Creating high pixel image

```python
1280*720
img = []
for i in range(720):
    temp = []
    for j in range(1280):
        temp.append(0)
        img.append(temp)
img = np.array(img)
```

# Exporting the image

```python
cv.imwrite('GRAY.png', img)
```
